# AGRUPAMENTO

```python
# quantos tipos de espécies? - retorna setosa, versicolor e virginica, 3 tipos
df["Species"].unique()
```

```python
# agrupamento dos 5 primeiros de cada espécie, setosa, versicolor e virginica
agrupar = df.groupby('Species')
agrupar.head()
```

```python
# maximo valor dentre as 3 espécies de largura, comprimento e pétala.comp
maximo = df.groupby('Species').max()
maximo
```

```python
# mínimo valor dentre as 3 espécies
minimo = df.groupby('Species').min()
minimo
```
