```python
sb.countplot(x='payment',data=base)
```

```python
# grafico de barras
agrup = base.groupby(['payment']).size()
agrup.plot.bar(color = 'green', ylabel = 'Quant')
```

```python
# grafico de pizza
agrup = base.groupby(['payment']).size()
agrup.plot.pie(legend=False,colors=['blue','red'],labels=['Cash','Credit Card'],fontsize=20, autopct='%.2f',table=False,ylabel='')
```
