## **Conhecendo a distribuição normal**
# Para descobrir se uma distrubuição é normal podemos utilizar o histograma, o box-plot, o gráfico QQ e/ou realizando teste de hipótese como de Shapiro



```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sb
from scipy import stats
```

```python
base = pd.read_csv('iris.csv')
```

```python
base
```

```python
sb.boxplot(base['Petal.Length']).set_title('Largura de Pétala')
```

```python
# teste de hipótese
# H0: considera a distribuição dos dados como normal
# H1: não é uma distribuição normal
# o valor p precisa ser maior que 0,05 (nivel de significância) para ser uma distribuição normal
shapiro_teste = stats.shapiro(base['Petal.Length'])
print(shapiro_teste.pvalue<=0.05)
# retorna: true, ou seja, menor q 0,05, não é uma distribuição normal
```

```python
sm.qqplot(base['Petal.Width'], color = "green", line = "45")
plt.show()
```
