

##  **Numpy**



```python
import numpy as np
```

```python
# usar a função np.array e passar uma lista para ele
a = np.array([1, 2, 3])
print(a)
```

```python
# montar uma matriz
matriz = np.array([[1,2],[3,4],[5,6]])
print(matriz)
```

```python
# tamanho da matriz
matriz.shape
```

```python
# com um intervalo de elementos utilizando a função arange(), sendo q inicia no valor zero e o último número nao é incluido no intervalo
a = np.arange(10)
print(a)
```

```python
# somar
a.sum()
```

```python
# valor máximo
print("Valor Máximo", a.max())
# valor mínimo
print("Valor Mínimo", a.min())
# média
print("Média", a.mean())
# somar todos os elementos com o valor 2
print("Soma com o valor acrescentado 2", (a+2))
```

# há duas estruturas de dados do panda chamados de Series(uma dimensão) e o DataFrame(duas dimensões)
# Os DataFrames, também, podem ser criados a partir de listas, tuplas, dicionários, arquivos externos.

```python
df = pd.DataFrame(
    {
        "Nome": [
            "Luiz Silva",
            "Maria Luisa da Silva",
            "Roberto Silva",
        ],
        "Idade": [22, 35, 58],
    }
)
df
```

```python
# Selecionando uma Série
df["Nome"]
```

```python
# Criando uma Série
idades = pd.Series([22, 35, 58], index=["Aluno 1", "Aluno 2","Aluno 3"])
idades
```
