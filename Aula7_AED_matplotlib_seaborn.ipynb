```python
import matplotlib.pyplot as plt
import seaborn as srn
import pandas as pd
```

```python
# grafico de barras vertical
linguagens = ['C','C++','SQL','HTML','JAVA','PYTHON']
estudantes = [23, 17, 35, 29, 12, 40]
plt.bar(linguagens,estudantes)
plt.xlabel('Cursos')
plt.ylabel('Estudantes')
plt.title('Alunos matriculados por curso')
plt.show()
```

```python
# grafico de barras horizontal
linguagens = ['C','C++','SQL','HTML','JAVA','PYTHON']
estudantes = [23, 17, 35, 29, 12, 40]
plt.barh(linguagens,estudantes)
plt.xlabel('Estudantes')
plt.ylabel('Cursos')
plt.title('Alunos matriculados por curso')
plt.show()
```
