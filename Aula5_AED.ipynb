# Analisar uma variável peso

```python
# amplitude de idades
minimo = base['body_mass_g'].min()
maximo = base['body_mass_g'].max()
print("Amplitude",maximo-minimo)
```

```python
media = base['body_mass_g'].mean()
media
```

```python
# observar a média de pesos considerando o sexo
media = base.groupby(['sex'])['body_mass_g'].mean()
media
```

```python
# visualizar a moda(valor q ocorre com mais frequência)
moda = base['body_mass_g'].mode()
moda
```

```python
# mediana: divide um conjunto de dados ordenados em duas partes. esse valor ocupa a posição do meio(central) do conjunto
mediana = base['body_mass_g'].median()
mediana
# retorna 4050.0 ou seja, este é o valor da posição central dos pesos, 50% menor q este valor, 50% maior
```

```python
# quartis: dividem um conjunto em 4 partes, quartos, primeiro quartil = 25%
# segundo, 50, terceiro 75 e 4 100%
quartil1 = base['body_mass_g'].quantile(q=0.25)
quartil1
```

```python
# segundo quartil, como é 50% é o mesmo valor da mediana
quartil2 = base['body_mass_g'].quantile(q=0.50)
quartil2
```

```python
# terceiro quartil, 75%
quartil3 = base['body_mass_g'].quantile(q=0.75)
quartil3
```

```python
# quarto quartil, mesma coisa q o valor máximo(max)
quartil4 = base['body_mass_g'].quantile(q=1.00)
quartil4
```
