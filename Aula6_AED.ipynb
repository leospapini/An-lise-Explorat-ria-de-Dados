```python
import pandas as pd
```

```python
base = pd.read_csv('penguins.csv')
```

```python
base
```

```python
# desvio médio absoluto: trata-se da diferença entre um valor da média
base.body_mass_g.mad()
```

```python
# variância
base.body_mass_g.var()
```

```python
# desvio padrão
base.body_mass_g.std()
```

```python
# coeficiente de variação
# analisar a dispersão em termos relativos, útil na comparação de variáveis
coef = base.body_mass_g.std()/base['body_mass_g'].mean()*100
coef
```

```python
# contagem, media, desvio padrao, minimo, quartis e  maximo
base['body_mass_g'].describe()
```
