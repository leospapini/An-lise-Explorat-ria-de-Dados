```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
cursoSexoAlunos = pd.DataFrame ([[100,20],
                                 [40,40]],
                                index = ['Engenharia', 'Psicologia'],
                                columns = ['Masculino','Feminino'])
cursoSexoAlunos
```

```python
cursoSexoAlunos.plot.bar(stacked=True)
```

# **Teste Qui-Quadrado:** teste de hipótese estatística que assume que as frequências observadas para uma variável categórica correspondem as frequências esperadas para uma variável categórica
# **H0: se for aceita significará que não há associação entre as variáveis**
# **H1: as frequencias observadas e esperadas são diferentes entre as variáveis associadas**
# Para **rejeitar H0 o pvalor (probabilidade de teste) é menor que o nível de significância**

```python
c_sa,p,dof,expected = chi2_contingency(cursoSexoAlunos)
```

```python
p<0.05
# como é menor então rejeitamos H0, ou seja, as variáveis são associadas
```
